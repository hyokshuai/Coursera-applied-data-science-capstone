# Get the Data and do Data Cleaning

```python
import bs4 as bs
#Getting the data from url as table
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
res = requests.get(url)
soup = bs.BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))
data = pd.read_json(df[0].to_json(orient='records'))
data_useful = data[data['Borough']!= 'Not assigned']
data_useful = data[data['Neighbourhood']!= 'Not assigned']
data_useful=data_useful.reset_index(drop=True)
name_list=['M5G','M2H','M4B','M1J','M4G','M4M','M1R','M9V','M9L','M5V','M1B','M5A']
new=pd.DataFrame(columns=['Postal Code','Borough','Neighbourhood'])
a=[]
for i in range(0,12):
    a.append(name_list[i])
    new=new.append(data_useful[data_useful['Postal Code'].isin(a)])
    a=[]
print(data_useful.shape)
new.head(12).reset_index(drop=True)
```
